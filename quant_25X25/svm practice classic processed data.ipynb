```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df=pd.read_csv('prefinaloutput.csv')
```

```python
df.head(50)
```

```python
X=df.drop('isFraud',axis=1)
```

```python
X
```

```python
y=df['isFraud']
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
ts=0.5
X_train, X_test, y_train, y_test = train_test_split(X, y,stratify=y, test_size =ts)
```

```python
from sklearn.svm import SVC
svclassifier = SVC(C=5.0,kernel='poly',degree=5,gamma='scale')
svclassifier.fit(X_train, y_train)
```

```python
y_pred = svclassifier.predict(X_test)
```

```python
from sklearn.metrics import classification_report, confusion_matrix
print(confusion_matrix(y_test,y_pred))
print(classification_report(y_test,y_pred))
```

```python
from sklearn.metrics import f1_score
```

```python
cl=45
cs=15
cresult=[]
ci=[]
```

```python
for i in range(cs,cl+1):
    for j in range(5):
        X_train, X_test, y_train, y_test = train_test_split(X, y,stratify=y, test_size = ts)
        svclassifier = SVC(C=i,kernel='poly')
        svclassifier.fit(X_train, y_train)
        y_pred = svclassifier.predict(X_test)
        cresult.append(f1_score(y_test,y_pred))
        ci.append(i)
```

```python
plt.figure().set_figwidth(40)
plt.plot(cresult, linestyle = 'dotted')
plt.xticks(range(len(ci)), ci)
plt.xlabel("C parameter", labelpad=7)
plt.show()
```
