```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df=pd.read_csv('prefinaloutput.csv')
```

```python
df.head(50)
```

```python
X=df.drop('isFraud',axis=1)
```

```python
X
```

```python
y=df['isFraud']
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y,stratify=y, test_size = 0.25)
```

```python
from sklearn.svm import SVC
svclassifier = SVC(C=100.0,kernel='poly',degree=20,gamma='scale')
svclassifier.fit(X_train, y_train)
```

```python
y_pred = svclassifier.predict(X_test)
```

```python
from sklearn.metrics import classification_report, confusion_matrix
print(confusion_matrix(y_test,y_pred))
print(classification_report(y_test,y_pred))
```

```python
from sklearn.metrics import f1_score
```

```python
cl=45
cs=15
cresult=[]
ci=[]
```

```python
for i in range(cs,cl+1):
    for j in range(5):
        X_train, X_test, y_train, y_test = train_test_split(X, y,stratify=y, test_size = 0.25)
        svclassifier = SVC(C=i,kernel='poly')
        svclassifier.fit(X_train, y_train)
        y_pred = svclassifier.predict(X_test)
        cresult.append(f1_score(y_test,y_pred))
        ci.append(i)
```

```python
plt.figure().set_figwidth(40)
plt.plot(cresult, linestyle = 'dotted')
plt.xticks(range(len(ci)), ci)
plt.xlabel("C parameter", labelpad=7)
plt.show()
```

**THE GRID SEARCH FITTING**

```python
from sklearn.model_selection import GridSearchCV
from sklearn.pipeline import Pipeline
params = [{ 'C': [0.1,1,5,10, 100, 1000],  
   'gamma': [1, 0.1, 0.01, 0.001, 0.0001], 
   'kernel': ['rbf','linear','poly']  }]
```

```python
grid = GridSearchCV(SVC(), param_grid=params,  scoring='f1_macro', refit = True, verbose = 3,n_jobs=-1)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y,stratify=y, test_size = 0.25)
grid.fit(X_train, y_train)


print(grid.best_params_) 
grid_predictions = grid.predict(X_test)

print(classification_report(y_test, grid_predictions)) 
```
