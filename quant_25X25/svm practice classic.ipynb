```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df=pd.read_excel('Fraud_detection_Testdata2.1-1.xlsx', index_col=0)
```

```python
df.head(5)
```

```python
df=df.drop("Unnamed: 12",axis=1)
df=df.drop("IP Address/Location",axis=1)
df=df.drop("Date",axis=1)
df=df.drop("Time",axis=1)
df=df.drop("Comments ( Fraudulent txns )",axis=1)
```

```python
df.head(5)
```

```python
df=df.reset_index(drop=True)
```

```python
df.head(10)
```

```python
pd.Categorical(df['Instrument or Mode'])
```

```python
df['Instrument or Mode']=pd.Categorical(df['Instrument or Mode']).codes
```

```python
df.head(10)
```

```python
pd.Categorical(df['type'])
```

```python
df['type']=pd.Categorical(df['type']).codes
df['AccountNoOrig']=pd.Categorical(df['AccountNoOrig']).codes
df['AccountNoDest']=pd.Categorical(df['AccountNoDest']).codes
```

```python
df.head(5)
```

```python
X=df.drop('isFraud',axis=1)
```

```python
X
```

```python
y=df['isFraud']
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.85)
```

```python
from sklearn.svm import SVC
svclassifier = SVC(kernel='linear')
svclassifier.fit(X_train, y_train)
```

```python
y_pred = svclassifier.predict(X_test)
```

```python
from sklearn.metrics import classification_report, confusion_matrix
print(confusion_matrix(y_test,y_pred))
print(classification_report(y_test,y_pred))
```
